```r
setwd("~/Projects/CommonG/Tiles/")
library("ggplot2")
```

```r
tiles_raw <- read.csv("Oyster Tile Size Data - 4_18_16.csv",header = TRUE)
```

```r
tiles_raw
```

```r
pops <- gsub("[0-9]","",as.character(tiles_raw$Tile))
pops <- gsub("[AB]","",as.character(pops))
pops <- gsub("HC","Hood Canal",pops)
pops <- gsub("NF","Fidalgo Bay",pops)
pops <- gsub("SS","Oyster Bay",pops)
tiles_raw$Pop <- pops
tiles_raw_sort <- tiles_raw[order(tiles_raw$Pop),]
```

```r
tiles_raw_sort
```

```r
tiles_raw_sort$X10.14.mean <- sapply(c(1:nrow(tiles_raw_sort)), FUN = function(x, tr) {(tr$X10.14.length.A[x] +tr$X10.14.length.B[x])/2}, tr = tiles_raw_sort)
```

```r
ggplot(tiles_raw_sort,aes(Pop,X10.14.mean))+ labs(x="Population",y="Length (cm)") +geom_boxplot(aes(fill = Pop)) +ggtitle("Shell Length in Juvenile F2 Olympia Oysters\nOct. 14, 2015") +theme(plot.title = element_text(face="bold"))
```

```r
tiles_raw_sort$X8.27.mean <- sapply(c(1:nrow(tiles_raw_sort)), FUN = function(x, tr) {(tr$X8.27.length.A[x] +tr$X8.27.length.B[x])/2}, tr = tiles_raw_sort)
```

```r
ggplot(tiles_raw_sort,aes(Pop,X8.27.mean))+ labs(x="Population",y="Length (cm)") +geom_boxplot(aes(fill = Pop)) +ggtitle("Shell Length in Juvenile F2 Olympia Oysters\nAug. 27, 2015") +theme(plot.title = element_text(face="bold"))
```

```r
tiles_raw_sort$Aug_Octgrowth <- sapply(c(1:nrow(tiles_raw_sort)), FUN = function(x, tr) {(tr$X10.14.mean[x] - tr$X8.27.mean[x])/48}, tr = tiles_raw_sort)
tiles_raw_sort$Aug_Oct <- sapply(c(1:nrow(tiles_raw_sort)), FUN = function(x, tr) {tr$X10.14.mean[x] - tr$X8.27.mean[x]}, tr = tiles_raw_sort)
```

```r
ggplot(tiles_raw_sort,aes(Pop,Aug_Octgrowth))+ labs(x="Population",y="Growth Rate (cm/day)") +geom_boxplot(aes(fill = Pop)) +ggtitle("Growth Rate of Shell Length in Juvenile Olympia Oysters") +theme(plot.title = element_text(face="bold"))
```

```r
aov.growth <- aov(Aug_Octgrowth~Pop,data=tiles_raw_sort)
summary(aov.growth)
```

```r
aov.Oct <- aov(X10.14.mean~Pop,data=tiles_raw_sort)
summary(aov.Oct)
```

```r
aov.Aug <- aov(X8.27.mean~Pop,data=tiles_raw_sort)
summary(aov.Aug)
```

```r
Oct_ptest <- pairwise.t.test(tiles_raw_sort$X10.14.mean,tiles_raw_sort$Pop,p.adjust = "bonferroni")
Oct_ptest
```

```r
Aug_ptest <- pairwise.t.test(tiles_raw_sort$X8.27.mean,tiles_raw_sort$Pop,p.adjust = "bonferroni")
Aug_ptest
```

```r
Growth_ptest <- pairwise.t.test(tiles_raw_sort$Aug_Octgrowth,tiles_raw_sort$Pop,p.adjust = "bonferroni")
Growth_ptest
```
